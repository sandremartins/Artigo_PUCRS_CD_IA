Os alunos usam o software de visualização de dados do Tableau para descobrir insights sobre o caso. Este é um conjunto de dados sintéticos criado especificamente para acompanhar o estudo de caso.

Todos os anos, mais ou menos, atualizamos o conjunto de dados para incluir colunas adicionais e fazer pequenas alterações nos dados subjacentes. Nesta versão, fizemos as seguintes atualizações:
